```python
import os
import pandas as pd

# Path to the folder containing the extracted CSV files
folder_path = 'training_data'  # Change to your folder path if different

# List all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Read and display each CSV file
for file in csv_files:
    file_path = os.path.join(folder_path, file)  # Get the full path of the file
    df = pd.read_csv(file_path)  # Read the CSV file into a DataFrame
    print(f"Contents of {file}:")
    print(df.head())  # Display the first few rows of the CSV file
    print()  # Add a newline for readability
```
