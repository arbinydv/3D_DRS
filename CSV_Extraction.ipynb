```python
import zipfile
import os

# Path to the zip file (in the same directory as the script)
zip_file_path = 'digits_3d_training_data.zip'  # Replace with the actual zip file name
# New folder to extract CSV files
new_folder = 'extracted_csv_files/'

# Create the folder if it doesn't exist
os.makedirs(new_folder, exist_ok=True)

# Unzip the folder and extract only CSV files
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Get list of all files in the zip
    all_files = zip_ref.namelist()
    
    # Extract only CSV files
    csv_files = [f for f in all_files if f.endswith('.csv')]

    # Extract CSV files to the new folder
    for file in csv_files:
        zip_ref.extract(file, new_folder)
        print(f"Extracted {file} to {new_folder}")

print(f"CSV files have been extracted to {new_folder}")
```
